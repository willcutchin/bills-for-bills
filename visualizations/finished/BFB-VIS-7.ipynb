```python
import pandas as pd
import matplotlib.pyplot as plt
import requests
import io
import seaborn as sns
import matplotlib.ticker as mtick
```

```python
#url variables
politician_data_csv_url = "https://raw.githubusercontent.com/willcutchin/bills-for-bills/main/data-sets/politician-trading-data/politician-trading-data-csv/politician-trading-data-BUY2020-final.csv"

#download csv
pol_download = requests.get(politician_data_csv_url).content
```

```python
#read file to data frame
pol_df = pd.read_csv(io.StringIO(pol_download.decode('utf-8')))
```

```python
pol_df["TRANSACTION_DATE"] = pd.to_datetime(pol_df["TRANSACTION_DATE"])
pol_df_1 = len(pol_df[pol_df["TRANSACTION_DATE"].dt.month == 1])
pol_df_2 = len(pol_df[pol_df["TRANSACTION_DATE"].dt.month == 2])
pol_df_3 = len(pol_df[pol_df["TRANSACTION_DATE"].dt.month == 3])
pol_df_4 = len(pol_df[pol_df["TRANSACTION_DATE"].dt.month == 4])
pol_df_5 = len(pol_df[pol_df["TRANSACTION_DATE"].dt.month == 5])
pol_df_6 = len(pol_df[pol_df["TRANSACTION_DATE"].dt.month == 6])
pol_df_7 = len(pol_df[pol_df["TRANSACTION_DATE"].dt.month == 7])
pol_df_8 = len(pol_df[pol_df["TRANSACTION_DATE"].dt.month == 8])
pol_df_9 = len(pol_df[pol_df["TRANSACTION_DATE"].dt.month == 9])
pol_df_10 = len(pol_df[pol_df["TRANSACTION_DATE"].dt.month == 10])
pol_df_11 = len(pol_df[pol_df["TRANSACTION_DATE"].dt.month == 11])
pol_df_12 = len(pol_df[pol_df["TRANSACTION_DATE"].dt.month == 12])
```

```python
month_freq_data = {"MONTH": ['January','February','March','April','May','June','July','August','September','October','November','December'], 
                   'TRADE_FREQ': [pol_df_1,pol_df_2,pol_df_3,pol_df_4,pol_df_5,pol_df_6,pol_df_7,pol_df_8,pol_df_9,pol_df_10,pol_df_11,pol_df_12]}
```

```python
pol_month_freq = pd.DataFrame(data=month_freq_data)
pol_month_freq['TRADE_FREQ'] = pd.to_numeric(pol_month_freq['TRADE_FREQ'], downcast="integer")
pol_month_freq.head()
```

```python
palette = sns.color_palette("crest")
ax = sns.lineplot(x="MONTH",y="TRADE_FREQ",data=pol_month_freq,color='seagreen')
ax.set_xlabel("Month of Trading",fontsize=20,labelpad=20)
ax.set_ylabel("Number of Trades",fontsize=20,labelpad=20)
ax.set_title("Trade Frequency by Month (2020)", fontsize=20, pad=40)
ax.axvline("March",color="darkblue",linestyle="--",lw=2)
ax.text("April", 1200,"Passing of CARES Act (blue --)")
plt.xticks(rotation=90)
```
